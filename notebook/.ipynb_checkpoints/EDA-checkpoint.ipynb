```python
# toDate (format='%d-%b-%y')
def df_date(df, columns):
    for col in columns:
        df[col] = pd.to_datetime(df[col], format='%d-%b-%y')
    return df

# toDate - ApprovalFY contains some letter
df['ApprovalFY'] = df['ApprovalFY'].str.extract('(\d+)', expand=False)
df['ApprovalFY'] = pd.to_datetime(df.ApprovalFY, format='%Y')

# toDate
df = df_date(df, columns=['ApprovalDate', 'DisbursementDate', 'ChgOffDate'])
```

```python
# extract only number
def only_num(df, columns):
    for col in columns:
        df[col] = df[col].str.extract('(\d+)', expand=False).astype(int)
    return df

df = only_num(df, columns=['DisbursementGross', 'BalanceGross', 'ChgOffPrinGr', 'GrAppv', 'SBA_Appv'])
```

```python
# continous function plot
def plot_con(graphs_col, dataframe, figsize):
    graphs_col = graphs_col
    graphs_row = (len(dataframe.columns) // graphs_col) + (len(dataframe.columns) % graphs_col > 0)

    fig = plt.figure(figsize=figsize)
    for i in range(1, len(dataframe.columns)+1):
        ax = fig.add_subplot(graphs_row, graphs_col, i)
        sns.histplot(x=dataframe.iloc[:, i-1], kde=True, ax=ax)
        plt.tight_layout()
```

```python
plot_con(graphs_col=3, dataframe=df, figsize=(12,5))
```
